```python
column_heading = []
entity_val = []
for i in unchanged_df.columns:
    if 'RIL' in str(i):
        result = "_".join(i.split("_")[1:])
        column_heading.append(result)
    result1 = "_".join(str(i).split("_")[:1])
    if result1 not in entity_val:
        entity_val.append(result1)
```

```python
table_to_change = unchanged_df.drop(['year', 0, '0.1'], axis=1)
```

```python
list_table_to_change = table_to_change.values.tolist()
```

```python
# original_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

transposed_list_table = list(map(list, zip(*list_table_to_change)))

```

```python
len(transposed_list_table)
```

```python
from itertools import chain

matrix2 = []
for k1 in range(22):
    values = []
    for i in range(k1, len(transposed_list_table), 22):
        values.append(transposed_list_table[i])
    flattened_list = list(chain.from_iterable(values))
    matrix2.append(flattened_list)
```

```python
final_table = list(map(list, zip(*matrix2)))
```
